```python
diet_categories = []
instructions_text = []
ingredients = []

for recipe in recipes:
    recipe_ingredients = []
    diet_categories.append(itemgetter('vegetarian', 'vegan', 'glutenFree', 'dairyFree', 'veryHealthy', 'cheap', 
                                       'veryPopular', 'sustainable', 'weightWatcherSmartPoints', 'gaps', 'lowFodmap', 
                                       'ketogenic', 'whole30')(recipe))
    instructions_text.append(recipe['instructions'])
    for ingr_dict in recipe['extendedIngredients']:
        recipe_ingredients.append(ingr_dict['originalName'])
    ingredients.append(recipe_ingredients)
        
    
```
